```python
setosa = data[:50]
print(setosa)
```

```python
versicolor = data[50:100]
print(versicolor)
```

```python
virginica = data[100:]
print(virginica)
```

```python
plt.title("Iris Setosa")
plt.hist([setosa["SepalLengthCm"],setosa["SepalWidthCm"],setosa["PetalLengthCm"],setosa["PetalWidthCm"]],label = ["sepal length","sepal width","petal length","petal width"])
plt.legend()
```

```python
plt.title("Iris versicolor")
plt.hist([versicolor["SepalLengthCm"],versicolor["SepalWidthCm"],versicolor["PetalLengthCm"],versicolor["PetalWidthCm"]],label = ["sepal length","sepal width","petal length","petal width"])
plt.legend()
```

```python
plt.title("Iris virginica")
plt.hist([virginica["SepalLengthCm"],virginica["SepalWidthCm"],virginica["PetalLengthCm"],virginica["PetalWidthCm"]],label = ["sepal length","sepal width","petal length","petal width"])
plt.legend()
```

```python
colors = {'Iris-setosa':'red','Iris-versicolor':'green','Iris-virginica':'blue'}
plt.scatter(data["SepalLengthCm"],data["SepalWidthCm"],c=data['Species'].map(colors))
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
```

```python
plt.scatter(data["SepalLengthCm"],data["PetalLengthCm"],c=data['Species'].map(colors))
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
```

```python
plt.scatter(data["SepalLengthCm"],data["PetalWidthCm"],c=data['Species'].map(colors))
plt.xlabel("Sepal Length")
plt.ylabel("Petal Width")
```

```python
plt.scatter(data["SepalWidthCm"],data["PetalWidthCm"],c=data['Species'].map(colors))
plt.xlabel("Sepal Width")
plt.ylabel("Petal Width")
```

```python
plt.scatter(data["SepalWidthCm"],data["PetalLengthCm"],c=data['Species'].map(colors))
plt.xlabel("Sepal Width")
plt.ylabel("Petal Length")
```

```python
plt.scatter(data["PetalLengthCm"],data["PetalWidthCm"],c=data['Species'].map(colors))
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
```

```python
ax = plt.axes(projection ="3d")
ax.scatter3D(data["SepalLengthCm"],data["SepalWidthCm"],data["PetalLengthCm"],c=data['Species'].map(colors))
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Sepal Width")
ax.set_zlabel("Petal Length")
```

```python
ax = plt.axes(projection ="3d")
ax.scatter3D(data["PetalLengthCm"],data["PetalWidthCm"],data["SepalWidthCm"],c=data['Species'].map(colors))
ax.set_xlabel("Petal Length")
ax.set_ylabel("Petal Width")
ax.set_zlabel("Sepal Width")
```
